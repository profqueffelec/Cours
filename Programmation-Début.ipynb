```python
def facto(n):
    test=n
    accu=1
    while (test!=0):
        accu*=test
        test=test-1
    return accu
facto(6),facto(5)
```

```python
def facto(n):
    if n==0:
        return 1
    else:
        return n*facto(n-1)

facto(5),facto(4)
```

Variables locales     
$\rightarrow$ cessent d'exister en dehors de la fonction

```python
def euclide(a,b):
    quotient=0
    reste=a
    while(reste>=b):
        quotient+=1
        reste-=b
        #print(reste)
    return reste,quotient

#euclide(13,5)
#reste
```
